```python
import cv2
import numpy as np
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
from imutils import paths
import imutils
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
def getGray(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(r"C:\Users\Daniel Adama\Desktop\PROGRAMMING\1.3PythonDatascience\datascience\FaceDetection\OCR\Preprocessed_image\{}.jpg",image)
    return image

def getBlur(image) :
    img_blur = cv2.GaussianBlur(image, (7, 7),0)
    cv2.imwrite(r"C:\Users\Daniel Adama\Desktop\PROGRAMMING\1.3PythonDatascience\datascience\FaceDetection\OCR\Preprocessed_image\{}.jpg",img_blur)    
    return img_blur

# threshold
def getThresh(image):
    #pixels with value below 100 are turned black (0) and those with higher value are turned white (255)
    thresh_img = cv2.threshold(image, 100, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY)[1]    
    cv2.imwrite(r"C:\Users\Daniel Adama\Desktop\PROGRAMMING\1.3PythonDatascience\datascience\FaceDetection\OCR\Preprocessed_image\{}.jpg",thresh_img)
    return thresh_img

#path to input data
data_path = r'C:\Users\Daniel Adama\Desktop\PROGRAMMING\1.3PythonDatascience\datascience\FaceDetection\OCR\dataset'

pathToimages = list(paths.list_images(data_path))
#grab the images and preprocess it
for pathToimage in pathToimages:
    #preprocess the image
    image = cv2.imread(pathToimage)
    #get a copy of the image
    orig = image.copy()
    image = getGray(image)
    #image = imutils.resize(image, width=450)
    blur = getBlur(image)
    thresh = getThresh(blur)

    # find contours in the thresholded image and sort them by their size
    contours = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    contours = imutils.grab_contours(contours)
    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    
    # loop over the contours
    for cnt in contours:
        # compute the bounding box of the contour and use the contour to the bounding box
        (x, y, w, h) = cv2.boundingRect(cnt)
        
        #draw the rectangle on the copied image
        rect = cv2.rectangle(orig, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.imshow('Rectangle',rect)
        cv2.waitKey(0)
        
        roi = orig[y:y + h, x:x + w]
        
        #Apply Text detection 
        config = ('-l eng --oem 1 --psm 3')
        text = pytesseract.image_to_string(roi, config=config)
        words = ''.join(text)
        print('=='*3+ text +'=='*3)
        
        key = cv2.waitKey(5)
        if key == 27:
            break
```

```python
#save the text in a .txt file
with open(r'C:\Users\Daniel Adama\Desktop\PROGRAMMING\1.3PythonDatascience\datascience\FaceDetection\OCR\output.txt', 'w') as out:
    out.writelines(words)
```
